# Neural Networks

One of most common activation functions is the **sigmoid** function. This functions is mostly used as the last layer activation function.

$$sigmoid(x)=\frac{1}{1+e^{-x}}$$

$$sigmoid(x)'=sigmoid(x)\times(1-sigmoid(x))$$


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
  return 1 / (1 + np.exp(-x))
```

```python
def sigmoid_derivative(x):
  s = sigmoid(x)
  return s * (1 - s)
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
```

```python
plt.plot(x, y)
plt.title("Sigmoid Function")
plt.xlabel("x")
plt.ylabel("Sigmoid(x)")
plt.grid()
plt.show()
```

```python
y = sigmoid_derivative(x)
```

```python
plt.plot(x, y)
plt.title("Sigmoid Derivative")
plt.xlabel("x")
plt.ylabel("Sigmoid Derivative(x)")
plt.grid()
plt.show()
```

As you can see for very large or very small numbers the slope is very close to 0. Using this function as the activation function can slow down the gradient descent to find the optimom point.

## Feed Forward simulation

```python
x = np.array([[0.5], [0.3]])
w1 = np.array([[0.4, 0.7], [0.2, 0.9]])
b1 = np.array([[0.1], [-0.1]])
w2 = np.array([1.2, 0.5])
b2 = np.array([0.3])
```

```python
z1 = w1.dot(x) + b1
```

```python
a1 = np.empty((2, 1))
for i in range(2):
  a1[i][0] = sigmoid(z1[i][0])
```

```python
z2 = w2.dot(a1) + b2
```

```python
a2 = sigmoid(z2[0])
```
